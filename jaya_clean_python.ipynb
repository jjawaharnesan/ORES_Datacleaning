```python
import pandas as pd
import numpy as np
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
df = df.drop(['Daily Steps'], axis=1)
df['Sleep Duration'] = df['Sleep Duration'].astype(int)
df = df.rename(columns={'Gender': 'Sex'})
df['Sleep Disorder Risk Score'] = df['Sleep Duration'] + df['Quality of Sleep'] - df['Stress Level']
df = df.drop_duplicates()
lower_bound_value = 30
upper_bound_value = 60
df['Age'] = df['Age'].clip(lower=lower_bound_value, upper=upper_bound_value)
df[['systolic', 'diastolic']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)
df.to_csv('combined_sleep_health_dataset.csv', index=False)
df.head()
```
